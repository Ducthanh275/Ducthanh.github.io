```python
df2=np.array(df2)
num_of_cases, num_of_feature = df2.shape
test=np.split(df2, 4)
```

```python
class BS:
  def __init__(self, name, lat, lon, n, C):
    self.name = name
    self.lat = lat
    self.lon = lon
    self.n=n
    self.C=C
```

```python
BSloc=[]
for i in range (num_of_cases):
  Local=test[i].T
  BSloc.append(BS(Local[0],Local[1],Local[2],Local[3],Local[4]))
```

```python
df3=np.array(df3)
test2=np.split(df3.T, 6)
test2=np.array(test2)
c=[]
for i in range (test2.shape[0]):
  c.append(np.split(test2[i].T,60))
c=np.array(c)
c.shape[1]
c.shape
```

```python
name=np.array(name)
namecol=[]
for i in range (name.size):
  namecol.append(int(name[i]))
namecol=np.array(namecol)
```

```python
m=0
o=0
W1=[]
W2=[]
for q in range (c.shape[1]):
  B=[]
  dis=[]
  A_x1=[]
  A_x2=[]
  f=0
  for f in range (namecol.size-2):
    for j in range (num_of_cases):
      if (namecol[f+2]==BSloc[j].name):
        s=10**(((c[f+2][q]*BSloc[j].n)+BSloc[j].C)/(-10))
        dis=np.append(dis,s)
  dis=np.array(dis)
  dis=dis.T
  for i in range(x_gNB.size-1):
    A_x1.append(x_gNB[i+1]-x_gNB[0])
    A_x2.append(y_gNB[i+1]-y_gNB[0])
  A_x1=np.array(A_x1)
  A_x2=np.array(A_x2)
  A=np.column_stack((A_x1,A_x2))
  A=2*A

  b=x_gNB[0]**2+y_gNB[0]**2-dis[0]**2
  e=0
  B=[]
  for e in range(x_gNB.size-1):
    B.append((dis[0])**2-(dis[e+1])**2-(x_gNB[0])**2-(y_gNB[0])**2+(x_gNB[e+1])**2+(y_gNB[e+1])**2)
  B=np.array(B)
  B=B.T
  X=np.dot((np.dot(np.linalg.inv(np.dot(A.T,A)),A.T)),B)
  W2.append(lon0+X[0]*delta_x)
  W1.append(lat0+X[1]*delta_y)
W1=np.array(W1)
W2=np.array(W2)
W=np.column_stack((W1,W2))
W
```
